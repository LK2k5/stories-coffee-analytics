```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load clean datasets
monthly = pd.read_csv("../outputs/clean_monthly_sales.csv")
cat = pd.read_csv("../outputs/clean_category_summary.csv")
groups = pd.read_csv("../outputs/clean_group_sales.csv")
prod = pd.read_csv("../outputs/clean_product_profitability.csv")

# Quick check
monthly.head(), cat.head(), groups.head(), prod.head()
```

```python
import os

# Create folder if it doesn't exist
os.makedirs("../outputs/figures", exist_ok=True)
```

```python
# Filter for 2025
m2025 = monthly[monthly["year"] == 2025].copy()

# Top 10 branches
top10 = m2025.sort_values("total_calc", ascending=False).head(10)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.barplot(x="branch", y="total_calc", data=top10)

plt.xticks(rotation=75)
plt.title("Top 10 Branches by Total Sales (2025)")
plt.ylabel("Total Sales")
plt.xlabel("Branch")

plt.tight_layout()

plt.savefig("../outputs/figures/top10_branches_2025.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
plt.close()
```

```python
months = ["jan","feb","mar","apr","may","jun",
          "jul","aug","sep","oct","nov","dec"]

season = m2025[months].sum()

plt.figure(figsize=(10,5))

plt.plot(season.index, season.values, marker="o", linewidth=2)

plt.title("Total Sales by Month (2025)")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig("../outputs/figures/seasonality_2025.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate profit and revenue by branch
branch_margin = cat.groupby("branch")[["Total Profit", "Revenue"]].sum()

# Avoid division errors
branch_margin["Margin"] = np.where(
    branch_margin["Revenue"] != 0,
    branch_margin["Total Profit"] / branch_margin["Revenue"],
    0
)

# Sort descending
branch_margin = branch_margin.sort_values("Margin", ascending=False)

# Select top 15
top15 = branch_margin.head(15)

plt.figure(figsize=(12,5))

sns.barplot(
    x=top15.index,
    y=top15["Margin"]
)

plt.xticks(rotation=75)
plt.title("Top 15 Branches by Profit Margin (Food + Beverage)")
plt.ylabel("Profit Margin")
plt.xlabel("Branch")

plt.ylim(0, top15["Margin"].max() * 1.1)

plt.tight_layout()

plt.savefig("../outputs/figures/top_margins.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
plt.close()
```

```python
mix = cat.pivot_table(index="branch", columns="Category", values="Revenue", aggfunc="sum").fillna(0)
mix["Total"] = mix.sum(axis=1)
mix["BEV_share"] = mix.get("BEVERAGES",0)/mix["Total"]
mix["FOOD_share"] = mix.get("FOOD",0)/mix["Total"]

mix_sort = mix.sort_values("BEV_share", ascending=False).head(15)

plt.figure(figsize=(12,5))
sns.barplot(x=mix_sort.index, y=mix_sort["BEV_share"])
plt.xticks(rotation=75)
plt.title("Beverage Revenue Share (Top 15 branches by beverage share)")
plt.ylabel("Beverage share")
plt.tight_layout()
plt.savefig("../outputs/figures/bev_share.png", dpi=200)
plt.savefig("../outputs/figures/top_margins.png",
            dpi=300,
            bbox_inches="tight")
plt.show()
```
